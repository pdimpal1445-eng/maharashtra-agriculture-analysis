```python
import pandas as pd

df = pd.read_csv('rain.csv')
print(df.head())
print(df.shape)
print(df.columns)
```

```python
# Fakt Maharashtra data filter kara
maharashtra = df[df['State Name'] == 'maharashtra']
print(maharashtra.shape)
print(maharashtra.head())
```

```python
# Missing values kithe aahet?
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

# 1. Rice Production - Year wise chart
plt.figure(figsize=(12, 5))
plt.plot(maharashtra['Year'],
         maharashtra['RICE PRODUCTION (1000 tons)'],
         color='green', marker='o')
plt.title('Maharashtra Rice Production (1966-2017)')
plt.xlabel('Year')
plt.ylabel('Production (1000 tons)')
plt.grid(True)
plt.show()
```

```python
# Rain vs Rice Production relation
plt.figure(figsize=(12, 5))
plt.scatter(maharashtra['JUL'],
            maharashtra['RICE PRODUCTION (1000 tons)'],
            color='blue', alpha=0.6)
plt.title('July Rain vs Rice Production - Maharashtra')
plt.xlabel('July Rainfall (mm)')
plt.ylabel('Rice Production (1000 tons)')
plt.grid(True)
plt.show()
```

```python
# Onion Area - Year wise
plt.figure(figsize=(12, 5))
plt.plot(maharashtra['Year'],
         maharashtra['ONION AREA (1000 ha)'],
         color='red', marker='s')
plt.title('Maharashtra Onion Area (1966-2017)')
plt.xlabel('Year')
plt.ylabel('Area (1000 ha)')
plt.grid(True)
plt.show()
```

```python
# Simple AI Agent — Crop Recommendation
def crop_recommendation(rainfall):
    if rainfall < 100:
        return "🌵 Drought condition — Grow Millets or Sorghum"
    elif rainfall < 200:
        return "🌿 Low rainfall — Grow Wheat or Pulses"
    elif rainfall < 300:
        return "🌾 Good rainfall — Grow Rice or Sugarcane"
    else:
        return "🌧️ Heavy rainfall — Flood risk! Grow flood resistant crops"

# Test kara
print(crop_recommendation(243))  # Average Maharashtra rainfall
print(crop_recommendation(80))   # Minimum rainfall
print(crop_recommendation(392))  # Maximum rainfall
```

```python
# =======================================
# ADVANCED CROP RECOMMENDATION AI AGENT
# Maharashtra Farmer Helper 🌾
# =======================================

def advanced_crop_recommendation(rainfall, month):

    # Season identify kara
    if month in ['June', 'July', 'August', 'September']:
        season = 'Kharif (Monsoon)'
    elif month in ['October', 'November', 'December']:
        season = 'Rabi (Winter)'
    else:
        season = 'Summer'

    # Crop recommendation based on rainfall
    if rainfall < 100:
        crop = "Millets, Sorghum, Bajra"
        warning = "⚠️ Drought risk — Use drip irrigation!"
    elif rainfall < 200:
        crop = "Wheat, Chickpea, Pulses"
        warning = "✅ Low rainfall — Suitable for Rabi crops"
    elif rainfall < 300:
        crop = "Rice, Sugarcane, Onion"
        warning = "✅ Good rainfall — Best for Kharif crops!"
    else:
        crop = "Flood resistant Rice varieties"
        warning = "🌧️ Heavy rainfall — Flood risk! Be careful!"

    print(f"Season: {season}")
    print(f"Rainfall: {rainfall} mm")
    print(f"Recommended Crops: {crop}")
    print(f"Advisory: {warning}")
    print("-" * 45)

# Test with Maharashtra average data
print("=== Maharashtra Crop Recommendation Agent ===\n")

print("📍 Test 1 — Average July Rainfall:")
advanced_crop_recommendation(243, 'July')

print("📍 Test 2 — Drought Year:")
advanced_crop_recommendation(80, 'November')

print("📍 Test 3 — Heavy Rainfall Year:")
advanced_crop_recommendation(392, 'August')

print("📍 Test 4 — Winter Season:")
advanced_crop_recommendation(150, 'December')
```
